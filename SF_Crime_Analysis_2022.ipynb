as one of the safest places in SF is 80-85% safer than the other non safe neighborhoods

```python
# Load Dependecies and Settings
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
# Set path to the .csv file (Cleaned version as of 10/23/23)
file_path = Path("SF_Crime_Data_from_2018_to_Present_CLEANED.csv")

# print(file_path.exists())

# Read .csv
crimedata_cleaned = pd.read_csv(file_path)
crimedata_cleaned.head()
```

```python
# # Value counts for 'Analysis Neighborhood' column from 2018 to present
neighborhood_counts = crimedata_cleaned['Analysis Neighborhood'].value_counts()

# Print the value counts
print("Value Counts:")
print(neighborhood_counts)
print("\n")
```

```python
# Let's filter data for the year 2020
crimedata_cleaned_2020 = crimedata_cleaned[crimedata_cleaned['Incident Year'] == 2022]

# Value counts for 'Analysis Neighborhood' column for the year 2022
neighborhood_crime_2020 = crimedata_cleaned_2020['Analysis Neighborhood'].value_counts()
print(neighborhood_crime_2020)

# Mean of all the neighborhood for 2022
neighborhood_crime2020_mean = neighborhood_crime_2020.mean()
print(f"The average count of all crimes in 2020 is: {neighborhood_crime2020_mean}")
```
